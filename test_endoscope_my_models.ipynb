```python
# python test_model.py model=iphone_orig dped_dir=dped/ test_subset=full iteration=all resolution=orig use_gpu=true

from scipy import misc
from skimage import io
import numpy as np
import tensorflow as tf
from models import resnet
import utils
import os
import sys
from tensorflow.python.keras import backend as K

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
use_gpu = "true"
model_type = 'withoutNIMA'

model_dir = '/home/public/hw/project/DPED0424/models/withoutNIMA/'
model_name = 'my_withoutNIMA_iteration_19000.ckpt'

save_dir = '/home/public/hw/project/DPED0424/visual_results/endoscope/withoutNIMA/'

if not os.path.exists(save_dir):
    os.mkdir(save_dir)

test_dir = '/home/public/hw/dataset/endoscope/'

test_photos = [f for f in os.listdir(test_dir)]
config = tf.ConfigProto(device_count={'GPU': 0}) if use_gpu == "false" else None

for photo in test_photos:

    image  = misc.imread(os.path.join(test_dir,photo))
    IMAGE_HEIGHT, IMAGE_WIDTH, _ = image.shape
    IMAGE_SIZE = IMAGE_HEIGHT * IMAGE_WIDTH * 3
    image = np.float16(image)/255
    image_2d = np.reshape(image, [1, IMAGE_SIZE])


    with tf.Graph().as_default(), tf.Session(config=config) as sess:
        K.set_session(sess)

        # create placeholders for input images
        x_ = tf.placeholder(tf.float32, [None, IMAGE_SIZE])
        x_image = tf.reshape(x_, [-1, IMAGE_HEIGHT, IMAGE_WIDTH, 3])

            # generate enhanced image
        enhanced = resnet(x_image)

        saver = tf.train.Saver()
        saver.restore(sess, os.path.join(model_dir, model_name))

        print('Testing my final model, processing image %s'%(photo))
        enhanced_2d = sess.run(enhanced, feed_dict={x_:image_2d})
        enhanced_image = np.reshape(enhanced_2d, [IMAGE_HEIGHT, IMAGE_WIDTH, 3])

        before_after = np.hstack((image, enhanced_image))
        photo_name = photo.rsplit('.', 1)[0]

        # save the result
        misc.imsave(os.path.join(save_dir, model_type+'_'+photo_name+'_enhance.png'), enhanced_image)
        misc.imsave(os.path.join(save_dir, model_type+'_'+photo_name+'_before_after.png'), before_after)
```
